## Steps needed to use the pickle files stored from the model builder

```python
import pandas as pd
import pickle
```

```python
# Read in the cleaned csv generated by the analysis file
df = pd.read_csv(r'test_set_pipe.csv', keep_default_na=True, sep=',\s+', delimiter=',', skipinitialspace=True)
```

```python
# load the model from disk
# The file does show it, but the route number is 37 therefore we should use model 37
filename = "xgb_reg_model_37.pkl"
```

```python
#Load the model
with open(filename,'rb') as modelFile:
    model = pickle.load(modelFile)
```

```python
# Just to show the format of the data so we know which values we need
df
```

```python
#Predict with the test set
prediction = model.predict(df)
```

```python
# Returns value as an array but we will just extract that
# Value returned in seconds, we may want to convert to minutes/hours:minutes
# Actual journey time was 86 seconds, so it appears that the model is reasonably in line
# i.e. did not get corrupted during training
print(prediction)
```
